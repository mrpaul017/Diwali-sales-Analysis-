# Exploratory Data Analysis

## Gender

```python
#plotting a bar chart for gender
ax = sns.countplot(x='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#plotting a bar chart for Gender vs total amount
sales_gender = df.groupby(['Gender'], as_index=False)['Amount'].sum().sort_values(by='Amount',ascending=False)
sns.barplot(x='Gender', y='Amount', data=sales_gender);
```

*From the above graph we can see that most of the buyers are female*

## Age

```python
#plotting a bar chart for age groups
ax=sns.countplot(x='Age Group', hue='Gender', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
#Total amount vs Age group
sales_age = df.groupby(['Age Group'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False)
sns.barplot(x='Age Group', y='Amount', data= sales_age);
```

*From the above graph we can see that the most of the buyers are from age group of 26-35 years*

## State

```python
#Total number of orders from top 10 states
sales_states = df.groupby(['State'], as_index= False)['Orders'].sum().sort_values(by='Orders', ascending = False).head(10)
sns.set_theme(rc={'figure.figsize':(15,10)})
sns.barplot(x='State', y='Orders', data=sales_states);
```

```python
#Total amount of sales from top 10 states
sales_amount = df.groupby(['State'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False).head(10)
sns.set_theme(rc={'figure.figsize':(15,10)})
sns.barplot(x='State', y= 'Amount', data=sales_amount);
```

*From the above graphs we can see that most of the orders and amount of sales are from Uttar Pradesh, Maharastra, Karnataka respectively*

## Marital Status

```python
ax = sns.countplot(x='Marital_Status', data = df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_ms=df.groupby(['Marital_Status', 'Gender'], as_index=False)['Amount'].sum().sort_values(by= 'Amount', ascending = False)
sns.barplot(x='Marital_Status', y='Amount', hue='Gender', data=sales_ms);
sns.set_theme(rc={'figure.figsize':(8,5)})
```

*From the above graphs we can see that the number of buyers of unmarried women are higher than the rest and have higher purchasing power*

## Occupation

```python
sns.set_theme(rc={'figure.figsize':(20,15)})
ax= sns.countplot(x='Occupation',data=df)
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
sales_occup = df.groupby(['Occupation'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending = False )
sns.barplot(x='Occupation', y= 'Amount', data=sales_occup);
```

*From the above graph we can see most of the buyers from It Sector, Healthcare, Aviation sectors* 

## Product Category 

```python
ax=sns.countplot(x='Product_Category', data=df)
for bars in ax.containers:
    ax.bar_label(bars)
sns.set_theme(rc={'figure.figsize':(20,5)})
```

```python
sales_category = df.groupby(['Product_Category'], as_index=False)['Amount'].sum().sort_values(by='Amount', ascending= False).head(10)
sns.barplot(x='Product_Category', y='Amount', data = sales_category);
```

*From the above graph we can see most of the products sold are from Food, Clothing & Apparel, Electronics & Gadgets category*

## Conclusion

*Unmarried women age group 26-35 yrs from UP, Maharastra and Karnataka working in IT, Healthcare and Aviation are more likely to buy products from Food, Clothing & Apparel, Electronics & Gadgets category*
